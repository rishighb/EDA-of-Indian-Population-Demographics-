```python
# dividing the heirarchy


df_india = df[df['Level'] == 'India'].copy()


df_state = df[df['Level'] == 'STATE'].copy()


df_district = df[df['Level'] == 'DISTRICT'].copy()


df_sub_district = df[df['Level'] == 'SUB-DISTRICT'].copy()


print(f"Original Data Rows: {df.shape[0]}")
print(f"India Rows: {df_india.shape[0]}")
print(f"State Rows: {df_state.shape[0]}")
print(f"District Rows: {df_district.shape[0]}")
print(f"Sub-District Rows: {df_sub_district.shape[0]}")
```

```python


df_state = df[df['Level'] == 'STATE'].copy()
df_district = df[df['Level'] == 'DISTRICT'].copy()


df_state_total = df_state[df_state['TRU'] == 'Total'].copy()


df_state_ru = df_state[df_state['TRU'] != 'Total'].copy()


print(f"Total rows for ranking: {len(df_state_total)}")
print(f"Rural/Urban rows for comparison: {len(df_state_ru)}")
```

```python

#created metrics
def add_standard_metrics(data):
    df_calc = data.copy()
    
    
    df_calc['Literacy_Rate'] = (df_calc['P_LIT'] / df_calc['TOT_P']) * 100

    
    df_calc['Sex_Ratio'] = (df_calc['TOT_F'] / df_calc['TOT_M']) * 1000

    
    df_calc['Child_Sex_Ratio'] = (df_calc['F_06'] / df_calc['M_06']) * 1000

   
    df_calc['WPR'] = (df_calc['TOT_WORK_P'] / df_calc['TOT_P']) * 100
    
    return df_calc


df_state_metrics = add_standard_metrics(df_state_total)



#Urbanization Rate

subset_for_pivot = df_state[['Name', 'TRU', 'TOT_P']]

df_pivot = subset_for_pivot.pivot(index='Name', columns='TRU', values='TOT_P')

df_pivot['Urbanization_Rate'] = (df_pivot['Urban'] / df_pivot['Total']) * 100



df_final = df_state_metrics.merge(df_pivot[['Urbanization_Rate']], on='Name', how='left')


print("Feature Engineering Complete.")
print(df_final[['Name', 'Literacy_Rate', 'Sex_Ratio', 'Urbanization_Rate']].head())
```

```python

df_dist_stats = df_district[df_district['TRU'] == 'Total'].copy()

df_dist_stats['Literacy_Rate'] = (df_dist_stats['P_LIT'] / df_dist_stats['TOT_P']) * 100

# 3. Visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))


sns.histplot(df_dist_stats['TOT_P'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('District Population Distribution')
axes[0].set_xlabel('Total Population')
axes[0].set_ylabel('Frequency')


sns.histplot(df_dist_stats['Literacy_Rate'], kde=True, ax=axes[1], color='salmon')
axes[1].set_title('District Literacy Rate Distribution')
axes[1].set_xlabel('Literacy Rate (%)')
axes[1].set_ylabel('Frequency')

mean_lit = df_dist_stats['Literacy_Rate'].mean()
median_lit = df_dist_stats['Literacy_Rate'].median()

axes[1].axvline(mean_lit, color='green', linestyle='--', label=f'Mean: {mean_lit:.1f}%')
axes[1].axvline(median_lit, color='blue', linestyle='-', label=f'Median: {median_lit:.1f}%')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
def plot_state_rankings(data, metric, title_unit):
    sorted_df = data.sort_values(metric, ascending=False)
    
    top_5 = sorted_df.head(5)
    bottom_5 = sorted_df.tail(5)
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 5))
    
    sns.barplot(x=metric, y='Name', data=top_5, ax=axes[0], 
                palette='viridis', hue='Name', legend=False)
    
    axes[0].set_title(f'Top 5 States by {metric}')
    axes[0].set_xlabel(title_unit)
    axes[0].set_ylabel('')
    
    sns.barplot(x=metric, y='Name', data=bottom_5, ax=axes[1], 
                palette='magma', hue='Name', legend=False)
    
    axes[1].set_title(f'Bottom 5 States by {metric}')
    axes[1].set_xlabel(title_unit)
    axes[1].set_ylabel('')
    
    plt.tight_layout()
    plt.show()

plot_state_rankings(df_state_metrics, 'Literacy_Rate', 'Literacy Rate (%)')
plot_state_rankings(df_state_metrics, 'Sex_Ratio', 'Females per 1000 Males')
```

```python


df_sector = df_state_ru.copy()


df_sector['Total_CL'] = df_sector['MAIN_CL_P'] + df_sector['MARG_CL_P']


df_sector['Total_AL'] = df_sector['MAIN_AL_P'] + df_sector['MARG_AL_P']

df_sector['Total_HH'] = df_sector['MAIN_HH_P'] + df_sector['MARG_HH_P']


df_sector['Total_OT'] = df_sector['MAIN_OT_P'] + df_sector['MARG_OT_P']



df_sector['Cultivators'] = (df_sector['Total_CL'] / df_sector['TOT_WORK_P']) * 100
df_sector['Agri_Labor'] = (df_sector['Total_AL'] / df_sector['TOT_WORK_P']) * 100
df_sector['Household'] = (df_sector['Total_HH'] / df_sector['TOT_WORK_P']) * 100
df_sector['Other_Workers'] = (df_sector['Total_OT'] / df_sector['TOT_WORK_P']) * 100


sector_summary = df_sector.groupby('TRU')[['Cultivators', 'Agri_Labor', 'Household', 'Other_Workers']].mean().reset_index()

sector_melted = sector_summary.melt(
    id_vars='TRU',
    var_name='Sector',
    value_name='Percentage'
)

plt.figure(figsize=(10, 6))
sns.barplot(data=sector_melted, x='TRU', y='Percentage', hue='Sector', palette='Set2')

plt.title('Workforce Sector Distribution: Rural vs Urban')
plt.ylabel('Percentage of Workforce')
plt.xlabel('Area Type')
plt.legend(title='Economic Sector')
plt.tight_layout()
plt.show()
```
